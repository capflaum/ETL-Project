```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
diversity_file = "diversityindex.csv"
diversity_df = pd.read_csv(diversity_file)

states_file = "states_all.csv"
states_df = pd.read_csv(states_file)
states_df.head()
```

```python
states_df.columns
```

```python
states_8_df = states_df[['STATE', 'YEAR', 'ENROLL', 'TOTAL_REVENUE', 
                        'FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE',
                        'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE',
                        'SUPPORT_SERVICES_EXPENDITURE', 'OTHER_EXPENDITURE',
                        'CAPITAL_OUTLAY_EXPENDITURE','GRADES_8_G', 'GRADES_ALL_G',
                        'AVG_MATH_8_SCORE', 'AVG_READING_8_SCORE']]
```

```python
# Filter DataFrame by Year >= 2000
current_states_8_df = states_8_df.loc[states_8_df['YEAR'] >= 2000]
```

```python
# Fill NaN values
fillna_states_8_df = current_states_8_df.iloc[:, 2:].fillna(0.0)

# Convert Float to Int
int_states_8_df = fillna_states_8_df.iloc[:, 2:].astype(int)

# Join
non_int_states_8_df = current_states_8_df.iloc[:, :2]
current_states_8_df = non_int_states_8_df.join(int_states_8_df, how = "outer")
current_states_8_df.head()
```

```python
# Fix capitalizations
current_states_8_df.columns = current_states_8_df.columns.str.title()
current_states_8_df['State'] = current_states_8_df['State'].str.title()
current_states_8_df.head()
```

```python
connection_string = "postgres:Zumatower11!@localhost:5432/ETL_project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
current_states_8_df.to_sql(name='states', con=engine, if_exists='append', index=False)
```

```python
# Diversity File cleaning
diversity_df.head()
```

```python
diversity_index=diversity_df[['Location','Diversity-Index']]
diversity_index1 = diversity_index.rename(columns={"Location": "State", "Diversity-Index": "Diversity_Index"})
diversity_index1.head()
```

```python
diversity_state_df = diversity_index1[~diversity_index1["State"].str.contains(',')]
diversity_state_df.head()
```

```python
diversity_state_df.to_sql(name='diversity', con=engine, if_exists='append', index=False)
```

```python
query=f"""
    SELECT diversity."State", diversity."Diversity_Index", states."Total_Expenditure"
    FROM diversity
    INNER JOIN states on diversity."State"=states."State"
    WHERE states."Year"= 2013
    ORDER BY diversity."Diversity_Index"
    """
```

```python
diversity_expenditure = pd.read_sql_query(query, con=engine).head()
```
